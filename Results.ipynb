```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set_theme()
```

# Systematic literature review

```python
slr = pd.read_excel("SLR.xlsx")
```

```python
categories = slr["Category"].value_counts()
categories
```

```python
slr["Topic"].value_counts()#.to_excel("topics.xlsx")
```

```python
sentiment = slr["Sentiment"].value_counts()
sentiment
```

```python
slr.groupby(["Category", "Topic"])["Topic"].count()#.to_excel("categories_topics.xlsx")
```

```python
slr.groupby(["Topic", "Sentiment"])["Sentiment"].count()#.to_excel("topics_sentiment.xlsx")
```

```python
plt.figure(figsize=(7,4))

# create data
x = np.arange(1)
y1 = [23]
y2 = [20]
y3 = [7]
y4 = [0]
width = 0.01

# plot data in grouped manner of bar type
bar1 = plt.barh(x+0.01, y1, width, color="tab:blue")
plt.bar_label(bar1, padding=2, fontsize=13)
bar2 = plt.barh(x, y2, width, color="tab:green")
plt.bar_label(bar2, padding=2, fontsize=13)
bar3 = plt.barh(x-0.01, y3, width, color="tab:orange")
plt.bar_label(bar3, padding=2, fontsize=13)

plt.xlim(0, 50)
plt.xticks([0, 50], fontsize=14)
plt.yticks(x, ["Overall"], fontsize=14)
plt.xlabel("Frequency", fontsize=14)
plt.legend(["Application", "Implication", "Both"], fontsize=13)
plt.savefig("slrtypes.png", bbox_inches="tight", dpi=300)
plt.show()
```
